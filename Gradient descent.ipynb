```python
import  matplotlib.pyplot as plt
```

```python
x0 = 3#initialize paramater
lr = 0.01 #learning rate
precision = 0.000001 #this tells where algorithm to stop 
step_size = 1
iters = 0 #iterations counters 
max_iters = 10000 # max number of iterations
df = lambda x: 2*(x+5)#gradient of our function 
```

```python
lista = []
while step_size > precision and iters < max_iters:
    prev_x0 = x0
    x0 = x0 - (lr * df(prev_x0))
    step_size = abs(x0 - prev_x0)
    lista.append(step_size)
    iters += 1
    print("Iteration",iters,"\nX value is",x0)
print(lista)   
print(("The local minimum occurs at", x0))
```
