```python
#generate random numbers but do not keep the boundories for a and b 
mu, sigma = -0.0, 0.02
a, b = (-0.5 - mu) / sigma, (0.5 - mu) / sigma
values = truncnorm.rvs(a, b, scale=sigma, loc=mu, size=1000)
plt.hist(values, density = True, edgecolor = "black")
print(max(values))
```
